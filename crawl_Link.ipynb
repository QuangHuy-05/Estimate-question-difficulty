```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.edge.options import Options
from selenium.webdriver.common.by import By
import time
import json
```

```python
driver = webdriver.Chrome()
driver.get("https://vietjack.me/thi-online/lop-9")
time.sleep(2)
```

```python

subject_links = driver.find_elements(By.CSS_SELECTOR, "a.subject-item")
```

```python
# Lọc ra link môn Văn, Sử, Địa
target_subjects_XH = ['Văn', 'Lịch sử', 'Địa lý','Tiếng Anh (mới)']
subject_urls = {}
for link in subject_links:
    href = link.get_attribute("href")
    try:
        name_elem = link.find_element(By.CSS_SELECTOR, ".subject-item__name")
        name = name_elem.text.strip()
        if name in target_subjects_XH:
            subject_urls[name] = href
    except:
        continue
```

```python
from selenium.webdriver.common.by import By
# Tạo các "ma trận" lưu đề cho từng môn
matrix_van = []
matrix_su = []
matrix_dia = []
matrix_anh  = []
matrix_toan  = []
for subject, url in subject_urls.items():
    driver.get(url)
    print(f"\n----- Đề thi môn {subject} -----")
    time.sleep(3)
  
    test_blocks = driver.find_elements(By.CSS_SELECTOR, "p.test.m-b-10")
    subject_matrix = []

    for block in test_blocks:
        try:
            link = block.find_element(By.CSS_SELECTOR, "a.url-root")
            href = link.get_attribute("href")
            title = link.get_attribute("title")
            print(f"Title: {title}\n Link: {href}\n") 
            subject_matrix.append([title, href])
        except:
            continue  # Bỏ qua nếu không có link trong block đó
    
    if subject == "Văn":
        matrix_van = subject_matrix
    elif subject == "Lịch sử":
        matrix_su = subject_matrix
    elif subject == "Địa lý":
        matrix_dia = subject_matrix
    elif subject == "Tiếng Anh (mới)":
        matrix_anh = subject_matrix
    elif subject == "Toán":
        matrix_toan = subject_matrix
```

```python
import csv
import os

os.makedirs("output", exist_ok=True)

def save_matrix_to_csv(filename, matrix):
    with open(f"output/{filename}", mode="w", encoding="utf-8-sig", newline='') as f:
        writer = csv.writer(f)
        writer.writerow(["Title", "Link"])
        writer.writerows(matrix)

save_matrix_to_csv("van.csv", matrix_van)
save_matrix_to_csv("lich_su.csv", matrix_su)
save_matrix_to_csv("dia_ly.csv", matrix_dia)
save_matrix_to_csv("anhvan.csv", matrix_anh)
save_matrix_to_csv("toan.csv", matrix_toan)
print("Đã lưu xong các file CSV trong thư mục 'output'")
```
